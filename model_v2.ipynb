```python
from cmdstanpy import CmdStanModel
import pandas as pd
import src.etl as etl
```

```python
stan_file = "model_v2.stan"
model = CmdStanModel(stan_file=stan_file)
```

```python
raw_data = pd.read_csv("match_data.csv")
model_data, coords = etl.format_data_for_stan(input_df=raw_data)
```

```python
fit = model.sample(data=model_data)
```

```python
fit.summary()
```

```python
def extract_team_season_effects(fitted_model: CmdStanModel, coords: dict, off_param: str = "theta_off", def_param: str = "theta_def") -> pd.DataFrame:
    # coords: draws x teams x seasons

    ortg = fitted_model.stan_variable(off_param).mean(axis=0)
    drtg = fitted_model.stan_variable(def_param).mean(axis=0)

    full_df = pd.DataFrame()
    for season in range(len(coords['seasons'])):
        season_ortg = ortg[:,season]
        season_drtg = drtg[:,season]
        season_nrtg = season_ortg + season_drtg
        current_season = coords['seasons'][season]

        new_df = pd.DataFrame({'ortg' : season_ortg, 'drtg' : season_drtg, 'nrtg' : season_nrtg, 
                      'season' : current_season, 'team' : coords['teams']})
        
        full_df = pd.concat([full_df, new_df], ignore_index=True)

    full_df = full_df[['team', 'season', 'nrtg', 'ortg', 'drtg']].sort_values('nrtg', ascending=False)

    return full_df
```

```python
team_ratings = extract_team_season_effects(fitted_model=fit, coords=coords)
```

```python
efl_teams = raw_data.query('season == 2024 & league == "Championship"')['home_team'].unique()
```

```python
efl_team_ratings = team_ratings[team_ratings['team'].isin(efl_teams)].query('season == 2024')
print(efl_team_ratings.reset_index(drop=True))
```
